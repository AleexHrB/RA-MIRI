```python
import pandas as pd
import altair as alt
```

# First experiments

```python
books = ["crusoe", "dracula", "iliad", "mare-balena", "midsummer-nights-dream", "quijote", "valley-fear", "war-peace"]
reals = [6245, 9425, 8925, 5670, 3136, 23034, 5829, 17475]
hyperlog = []
record = []
hyper_std = []
record_std = [] 
hyper_SE = []
record_SE = []
for i in range(0,len(books)):
    df_hyper = pd.read_csv("./hyperlogBench/" + books[i] + ".txt", header=None)
    df_record = pd.read_csv("./recordBench/" + books[i] + ".txt", header=None)
    hyperlog.append(df_hyper.loc[:, 0].mean())
    record.append(df_record.loc[:, 0].mean())
    hyper_std.append(df_hyper.loc[:, 0].std())
    record_std.append(df_record.loc[:, 0].std())
    hyper_SE.append(hyper_std[i]/hyperlog[i])
    record_SE.append(record_std[i]/record[i])

full_df = pd.DataFrame( 
    {
        "Book": books,
        "HLL": hyperlog,
        "REC": record,
        "Real": reals,
        "HLL Deviation": hyper_std,
        "REC Deviation": record_std,
        "HLL SE": hyper_SE,
        "REC SE": record_SE,
    }
)
full_df
```

```python
dracula_hyper = pd.read_csv("./hyperlogBench/" + books[1] + ".txt", header=None)
dracula_record = pd.read_csv("./recordBench/" + books[1] + ".txt", header=None)
```

```python
dracula_hyper = dracula_hyper.rename(columns={0: "cardinality"})
dracula_record = dracula_record.rename(columns={0: "cardinality"})
```

```python

hist = alt.Chart(dracula_hyper).mark_bar().encode(
    alt.X('cardinality',bin = alt.BinParams(maxbins = 25), title="Estimated Cardinality"),
    y = 'count()',
).properties(
    title='Distribution of results of HyperLogLog for Dracula dataset'
)
hist.show()
```

```python

hist = alt.Chart(dracula_record).mark_bar().encode(alt.X('cardinality',
                                                       bin = alt.BinParams(maxbins = 25)),
                                             y = 'count()',
                                                 )

hist.show()
```

```python
alphas = [0.5*i for i in range(0,7)]
alphas[0] = 0
alphas[2] = 1
alphas[4] = 2
alphas[6] = 3
reals = [45 for i in range(0,7)]
hyperlog = []
record = []
hyper_std = []
record_std = [] 
hyper_SE = []
record_SE = []
for i in range(0,len(alphas)):
    df_hyper = pd.read_csv("./synthethicAlpha/syntData" + "{}".format(alphas[i]) + "SolutionHyper.txt", header=None)
    df_record = pd.read_csv("./synthethicAlpha/syntData" + "{}".format(alphas[i]) + "SolutionRecord.txt", header=None)
    hyperlog.append(df_hyper.loc[:, 0].mean())
    record.append(df_record.loc[:, 0].mean())
    hyper_std.append(df_hyper.loc[:, 0].std())
    record_std.append(df_record.loc[:, 0].std())
    hyper_SE.append(hyper_std[i]/hyperlog[i])
    record_SE.append(record_std[i]/record[i])

full_df = pd.DataFrame( 
    {
        "Alpha": alphas,
        "HLL": hyperlog,
        "REC": record,
        "Real": reals,
        "HLL Deviation": hyper_std,
        "REC Deviation": record_std,
        "HLL SE": hyper_SE,
        "REC SE": record_SE,
    }
)
full_df
```
