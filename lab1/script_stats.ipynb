```python
import pandas as pd
import altair as alt
```

```python

oneChoiceHundredBins = []
for i in range (1,5):
    n = 10**i
    oneChoiceHundredBins.append(pd.read_csv("./mandatoryExperiments/oneChoiceHundredBins/oneChoice" + f"{n}" + ".txt", header=None))

means = []
deviations = []
for i in range(1,5):
    means.append(oneChoiceHundredBins[i - 1].loc[:,0].mean())
    deviations.append(oneChoiceHundredBins[i - 1].loc[:,0].std())
means

hundred_df = pd.DataFrame(
    {
        "nBalls": [10**i for i in range(1,5)],
        "meanGap": means,
        "nBins": [100 for i in range(1,5)],
        "std": deviations,
    }
)
hundred_df.to_csv("./mandatoryExperiments/oneChoiceHundredBins/means.csv")
```

```python
oneChoiceThousandBins = []
for i in range (1,7):
    n = 10**i
    oneChoiceThousandBins.append(pd.read_csv("./mandatoryExperiments/oneChoiceThousandBins/oneChoice" + f"{n}" + ".txt", header=None))

means = []
deviations = []
for i in range(1,7):
    means.append(oneChoiceThousandBins[i - 1].loc[:,0].mean())
    deviations.append(oneChoiceThousandBins[i - 1].loc[:,0].std())
means

thousand_df = pd.DataFrame(
    {
        "nBalls": [10**i for i in range(1,7)],
        "meanGap": means,
        "nBins": [1000 for i in range(1,7)],
        "std": deviations,
    }
)
thousand_df.to_csv("./mandatoryExperiments/oneChoiceThousandBins/means.csv")
```

```python
alt.Chart(thousand_df).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white")).encode(
    x=alt.X('nBalls:N', title='Number of balls'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    tooltip= [alt.Tooltip('nBalls:N',title='Number of balls'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Thousand Bins'
)
```
