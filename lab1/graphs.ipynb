```python
import pandas as pd
import altair as alt
import math
```

# Graphs

## Base Experiments

```python
Choice_df = pd.read_csv("./mandatoryExperiments/globalMeans.csv")
Choice_df
```

```python
alpha = [math.log(x,y) for (x,y) in zip(Choice_df["nBalls"].tolist(), Choice_df["nBins"].tolist())]
Choice_alpha_df = Choice_df.copy()
Choice_alpha_df['alpha'] = alpha
Choice_alpha_df2 = Choice_alpha_df[Choice_alpha_df['alpha'] == 2.0]

MeanGapBeta = alt.Chart(Choice_alpha_df2).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white")).encode(
    x=alt.X('beta:Q', title='Beta value'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    color="nBins:N",
    tooltip= [alt.Tooltip('beta:Q',title='Value of Beta'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against Beta (alpha = 2)'
)

MeanGapBeta.display()
```

```python
oneChoice_df = Choice_df[Choice_df['beta'] == 0]
alpha = [math.log(x,y) for (x,y) in zip(oneChoice_df["nBalls"].tolist(), oneChoice_df["nBins"].tolist())]
oneChoice_df["alpha"] = alpha

MeanGapOne = alt.Chart(oneChoice_df).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white")).encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    color = "nBins:N",
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha (one choice)'
)

MeanGapOne.display()
```

```python
twoChoice_df = Choice_df[Choice_df['beta'] == 1]
alpha = [math.log(x,y) for (x,y) in zip(twoChoice_df["nBalls"].tolist(), twoChoice_df["nBins"].tolist())]
twoChoice_df["alpha"] = alpha

MeanGapTwo = alt.Chart(twoChoice_df).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white")).encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    color = "nBins:N",
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha (two choice)'
)

MeanGapTwo.display()
```

```python
m = 10000
ns = [m**(0.01*i) for i in range(25,200,1)]
alpha = [(0.01*i) for i in range(25,200,1)]
func = [math.log(math.log(ns[i])) for i in range(0,len(ns))]
func

func_df = pd.DataFrame(
    {
        "alpha": alpha,
        "function": func,
    }
)

functionTwo = alt.Chart(func_df).mark_line(color='red', strokeDash=[3,4]).encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('function:Q', title='Mean Gap'),
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha (two choice)'
)

twoChoice_df_10k = Choice_df[Choice_df['beta'] == 1][Choice_df['nBins'] == 10000]
alpha = [math.log(x,y) for (x,y) in zip(twoChoice_df_10k["nBalls"].tolist(), twoChoice_df_10k["nBins"].tolist())]
twoChoice_df_10k["alpha"] = alpha

MeanGapTwo10k = alt.Chart(twoChoice_df_10k).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white", color="red"), color="red").encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha (two choice)'
)
(MeanGapTwo10k + functionTwo) 
```

```python
m = 10000
ns = [m**(0.01*i) for i in range(25,200,1)]
alpha = [(0.01*i) for i in range(25,200,1)]
#func = [math.log(m)/math.log(math.log(m/ns[i])) for i in range(0,len(ns))]
func = [math.sqrt(ns[i]*math.log(m)/m)*1.5 for i in range(0,len(ns))]
func

func_df = pd.DataFrame(
    {
        "alpha": alpha,
        "function": func,
    }
)

functionOne = alt.Chart(func_df).mark_line(color="red", strokeDash=[3,4]).encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('function:Q', title='Mean Gap'),
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha (one choice)'
)

oneChoice_df_10k = Choice_df[Choice_df['beta'] == 0][Choice_df['nBins'] == 10000]
alpha = [math.log(x,y) for (x,y) in zip(oneChoice_df_10k["nBalls"].tolist(), oneChoice_df_10k["nBins"].tolist())]
oneChoice_df_10k["alpha"] = alpha

MeanGapOne10k = alt.Chart(oneChoice_df_10k).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white", color="red"), color="red").encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha (one choice)'
)
MeanGapOne10k + functionOne
```

```python
(MeanGapOne10k + functionOne) | (MeanGapTwo10k + functionTwo)
```

## Batched Experiments

```python
Batched_df = pd.read_csv("./batchedExperiments/generalMeans.csv")

alpha = [math.log(x,y) for (x,y) in zip(Batched_df["nBalls"].tolist(), Batched_df["nBins"].tolist())]
Batched_df['alpha'] = alpha
Batched_df
```

```python


batched_df_two = Batched_df[Batched_df["beta"] == 1]
batched_df_two_10k = batched_df_two[batched_df_two["nBins"] == 10000]

MeanGapTwo10k = alt.Chart(batched_df_two_10k).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white")).encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    color = alt.Color("batchPercentage:N",title='Batch size percentage'),
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha (two choice, m=10000)'
)

MeanGapTwo10k.display()
batched_df_two
```

```python


batched_df_one = Batched_df[Batched_df["beta"] == 0]
batched_df_one_10k = batched_df_one[batched_df_one["nBins"] == 10000]

MeanGapOne10k = alt.Chart(batched_df_one_10k).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white")).encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    color = alt.Color("batchPercentage:N",title='Batch size percentage'),
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha (one choice, m=10000)'
)

MeanGapOne10k.display()
```

```python


batched_df_two = Batched_df[Batched_df["beta"] == 1]
batched_df_two_10k = batched_df_two[batched_df_two["nBins"] == 10000]
batched_df_two_10k = batched_df_two_10k[batched_df_two_10k["batchPercentage"] != 0.5]

MeanGapTwo10k_no05 = alt.Chart(batched_df_two_10k).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white")).encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    color = alt.Color("batchPercentage:N",title='Batch size percentage'),
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha (two choice, m=10000)'
)

MeanGapTwo10k.display()
```

```python


batched_df_one = Batched_df[Batched_df["beta"] == 0]
batched_df_one_10k = batched_df_one[batched_df_one["nBins"] == 10000]
batched_df_one_10k = batched_df_one_10k[batched_df_one_10k["batchPercentage"] != 0.5]

MeanGapOne10k_strokeDash = alt.Chart(batched_df_one_10k).mark_line(point=alt.OverlayMarkDef(filled=True, fill="Purple", shape="diamond", size=150)).encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    color = alt.Color("batchPercentage:N",title='Batch size percentage'),
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha (one choice, m=10000)'
)

MeanGapOne10k_strokeDash.display()
```

```python
MeanGapOne10k_strokeDash | (MeanGapTwo10k_no05 + MeanGapOne10k_strokeDash)
```

```python
batched_df_beta = Batched_df[Batched_df["alpha"] == 2]
batched_df_beta = batched_df_beta[batched_df_beta["nBins"] == 10000]

betaChart = alt.Chart(batched_df_beta).mark_line(point=alt.OverlayMarkDef(filled=False, fill="White")).encode(
    x=alt.X('beta:Q', title='beta value'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    color = alt.Color("batchPercentage:N",title='Batch size percentage'),
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against betaa (alpha = 2, m=10000)'
)

betaChart
```

## Query

```python
query_df = pd.read_csv("./queryExperiments/generalMeans.csv")
query_df

alpha = [math.log(x,y) for (x,y) in zip(query_df["nBalls"].tolist(), query_df["nBins"].tolist())]
query_df['alpha'] = alpha

query_df
```

```python
query_1df = query_df[query_df["queries"] == 1]
MeanGapQuery = alt.Chart(query_1df).mark_line(point=alt.OverlayMarkDef(filled=True, size=50)).encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    color = "nBins:N",
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha'
)

MeanGapQuery.display()
```

```python
query_2df = query_df[query_df["queries"] == 2]
MeanGapQuery2 = alt.Chart(query_2df).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white"),strokeDash=[3,4]).encode(
    x=alt.X('alpha:Q', title='alpha'),
    y=alt.Y('meanGap:Q', title='Mean Gap'),
    color = "nBins:N",
    tooltip= [alt.Tooltip('alpha:Q',title='Alpha'), alt.Tooltip('meanGap:Q', title='Mean Gap')],
).properties(
    title='Mean Gap against alpha'
)

MeanGapQuery2.display()
```
